# LAZADA

```python
with open('lazadaID.json', encoding = 'utf-8') as f:
    data = [json.loads(line) for line in f]
data[0]
```

```python
lazada = pd.DataFrame(data= data, columns = ['id','conversation_id','created_at',  'date', 'tweet', 'likes_count','link'])
lazada
```

```python
lazada.info()
```

## Create Table Lazada

```python
mycursor.execute('''CREATE TABLE Lazada(
                                        id VARCHAR(30) NOT NULL,
                                        conversation_id VARCHAR(30) NOT NULL,
                                        created_at VARCHAR(20) NOT NULL,
                                        date DATE,
                                        tweet text,
                                        likes_count smallint,
                                        link VARCHAR(70)
                                        )
                ''')
```

```python
tb_lazada = '''INSERT INTO Lazada(id, conversation_id, created_at,  date, tweet, likes_count, link)
                VALUES (%s, %s, %s, %s, %s, %s, %s)
            '''
```

```python
for i,row in lazada.iterrows():
    mycursor.execute(tb_lazada, list(row))
mydb.commit()
```
